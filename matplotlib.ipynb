```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Valores do combstivel em R$')     # Titulo do grfico


plt.plot(gas['Year'], gas['Australia'],'r.-', label = 'Austrlia') # label serve para adicionar nomes dos paises, # r.- red (cores)
plt.plot(gas['Year'], gas['Canada'],'b.-',label = 'Canad')
plt.plot(gas['Year'], gas['France'],'m.-',label = 'France') 
plt.plot(gas['Year'], gas['Japan'],'g.-',label = 'Japo')
plt.plot(gas['Year'], gas['USA'],'y.-',label = 'USA')

plt.xticks(gas['Year'][::2])  # colocando no eixo x os valores de dois e dois anos.

plt.xlabel('Anos')  # legenda abaixo do grfico
plt.ylabel('R$')   # legenda ao lado do grfico


plt.legend() # colocando legenda
plt.show()   # print
```

```python
plt.title('Valores do combstivel em R$')     # Titulo do grfico

list_country = ['Australia', 'USA', 'Italy']

for country in list_country:
    plt.plot(gas['Year'], gas[country], label=[country], marker='.')
    
    plt.xticks(gas['Year'][::2])
```
